```python
import fastf1
import matplotlib.pyplot as plt
import os

# Create cache folder if missing
os.makedirs('cache', exist_ok=True)

# Enable FastF1 cache
fastf1.Cache.enable_cache('cache')

# Load session
session = fastf1.get_session(2024,"Suzuka","Q")
session.load()

# Get lap data
laps = session.laps
laps.pick_drivers("VER")
laps.pick_drivers("LEC")

# Pick fastest laps of two drivers
ver_fast = laps.pick_drivers("VER").pick_fastest()
lec_fast = laps.pick_drivers("LEC").pick_fastest()

# Get telemetry
ver_tel = ver_fast.get_car_data().add_distance()
lec_tel = lec_fast.get_car_data().add_distance()

ver_tel.head()
lec_tel.head()
```
